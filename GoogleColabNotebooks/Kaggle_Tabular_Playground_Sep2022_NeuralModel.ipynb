```python
import pandas as pd
import numpy as np
import tensorflow as tf
```

```python
data = pd.read_csv('/content/drive/MyDrive/ColabFiles/train.csv')
testData = pd.read_csv('/content/drive/MyDrive/ColabFiles/test.csv')
```

```python
year = []
month = []
datef = []

for i in data['date']:
    x = i.split('-')
    if int(x[0])==2017:
        year.append(0)
    elif int(x[0])==2018:
        year.append(1)
    elif int(x[0])==2019:
        year.append(2)
    elif int(x[0])==2020:
        year.append(3)
    month.append(int(x[1])/12)
    datef.append(int(x[2])/31)

data['year'],data['month'],data['datef'] = year,month,datef

dummyData = pd.get_dummies(data,columns=['country','store','product'])
dummyData = dummyData.drop('row_id',axis=1)

train_x = dummyData
train_x = train_x.drop('date',axis=1)
train_x.rename(columns={'datef':'date'},inplace=True)

train_y = train_x.pop('num_sold')
```

```python
display(train_x)
display(train_y)
```

```python
tyear = []
tmonth = []
tdatef = []

for i in testData['date']:
    x = i.split('-')
    tyear.append(0.4)
    tmonth.append(int(x[1])/12)
    tdatef.append(int(x[2])/31)

testData['year'],testData['month'],testData['datef'] = tyear,tmonth,tdatef

dummyData = pd.get_dummies(data,columns=['country','store','product'])
dummyData = dummyData.drop('row_id',axis=1)

test_x = dummyData
test_x = test_x.drop('date',axis=1)
test_x.rename(columns={'datef':'date'},inplace=True)

test_y = test_x.pop('num_sold')
```

```python
# from tensorflow.python import metrics
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(units=64,activation='relu',input_shape=[len(train_x.keys())]))
model.add(tf.keras.layers.Dense(units=64,activation='relu'))
model.add(tf.keras.layers.Dense(units=1,activation='relu'))

#optimizer = tf.keras.optimizers.RMSprop()

model.compile(loss='mse',optimizer='adam',metrics=[tf.keras.metrics.Accuracy(),'mae'])
```

```python
model.fit(x=train_x,y=train_y,epochs=10)
```

```python
model.evaluate(x=test_x,y=test_y)
```
